## Pixabay API

Для использования API Pixabay.com нужно быть зарегистарированным пользователем и получить **api_key**.

Чтобы получить токент нужно посетить страницу https://pixabay.com/api/docs/ и нажать **Sign Up** в секции **key**




```python
# сохраним api_key в переменную
API_KEY = "15659343-8a986b3b751bb35a80c68YY90"
```

```python
# запишем в переменную наш "поисковый запрос" для изображения
image_to_find = "kitten"

# Формируем url (адрес веб-сайта), по которому можно обращаться для получения списка изображений,
# соответствующих поисковому запросу
url = f"https://pixabay.com/api/?key={API_KEY}&q={image_to_find}&image_type=photo"

print(url)
```

```python
# импортируем библиотеку urllib.request
# библиотека urllib включает в себя большое количество вспомогательных компонент для работы с интернетом
# для удобства эти компоненты разделены в подбиблиотеки внутри библиотеки urllib
# здесь мы импортируем только нужную подбиблиотеку urllib.request - для выполнения HTTP запросов
import urllib.request

# выполняем HTTP запрос к API Pixabay
req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
print(req)
```

```python
response = urllib.request.urlopen(req)
print(response)
```

```python
# Преобразовать HTTP ответ в строку
# важно: функцию .read() можно позвать только один раз - второй раз "читать уже будет нечего"
# чтобы прочитать еще раз - нужно заново сделать HTTP запрос
data_string = response.read().decode()
```

```python
# напечатаем, что же получилось - выглядит не очень читаемо
print(f"type of data is: {type(data_string)}")
print(data_string)
```

```python
# на самом деле ответ Pixabay API приходит в формате json
# используем библиотеку json

import json

data = json.loads(data_string)
```

```python
# напечатаем что получилось
# теперь это уже питоновский "словарь" и с ним можно обращаться как со словарем
# например, доставать значения по ключам

# все еще выглядит не очень читабельно
print(data)
```

```python
# обратимся как со словарем - достанем список ключей
print(data.keys())
```

```python
# мы уже знаем про библиотеку pprint! - используем ее

from pprint import pprint

pprint(data)
```

```python
# импортируем библиотеку random
# эта библиотека облегчает работу со случайными числами
# в частности, помогает сгенерировать случайное число с нужными критериями
# или выбрать случайный элемент из списка
import random

# выбираем случайное изображение из списка
random_hit = random.choice(data["hits"])
pprint(random_hit)
```

```python
# достанем из этой большой структуры сам URL адрес изображения
image_url = random_hit["largeImageURL"]

print(image_url)
```

```python
# скачиваем само изображение
req = urllib.request.Request(image_url, headers={'User-Agent': 'Mozilla/5.0'})
image_data = urllib.request.urlopen(req).read()
```

```python
# посмотрим что там внутри image_data
# там бинарное представление изображения в jpg формате
print(image_data)
```

```python
# используем библиотеку IPython.display чтобы показать изображение прямо в jupyter notebook
from IPython.display import Image

Image(image_data, width=500)
```

```python
# сразу зовем готовую функцию для парсинга ответов в формате json - .json()
data = response.json()
```

```python
pprint(data)
```

```python
# по старой схеме выбираем случайное изображение и сразу url
random_hit = random.choice(data["hits"])
image_url = random_hit["largeImageURL"]

# и скачиваем изображение и сразу достаем содержимое из http ответа
image_data = requests.get(image_url).content
```
